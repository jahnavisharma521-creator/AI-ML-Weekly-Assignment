## Introduction
Smart assistant with routing, tools, logging and JSON output.

```python
import logging, re
logging.basicConfig(level=logging.INFO)
```

```python
def calculator(expression):

    try:
        return eval(expression)
    except:
        return 'Invalid mathematical expression'

def extract_keywords(text):
    stop={'is','a','the','of','and','to','in','on','for','with'}
    words=re.findall(r'\\w+', text.lower())
    return [w for w in words if w not in stop]

def word_counter(text):
    return len(text.split())

def summarize_text(text):
    s=text.split('.')
    return '. '.join(s[:2])+'.' if len(s)>2 else text
```

```python
def agent(query):
    q=query.lower()
    if any(x in q for x in ['calculate','+','-','*','/']):
        return {'type':'calculation','tool':'calculator','result':calculator(q.replace('calculate','').strip())}
    elif 'keyword' in q:
        return {'type':'keyword_extraction','tool':'keyword_tool','result':extract_keywords(q.replace('extract keywords from',''))}
    elif 'count words' in q:
        return {'type':'word_count','tool':'word_counter','result':word_counter(q.replace('count words',''))}
    elif 'summarize' in q:
        return {'type':'summarization','tool':'summarizer','result':summarize_text(query.replace('Summarize','').replace('summarize',''))}
    else:
        return {'type':'general','tool':'none','result':'This is a general query. No specific tool was required.'}
```

```python
queries=['Calculate 25 + 10','Extract keywords from Artificial Intelligence is transforming industries','Count words Machine learning is very powerful','Summarize AI is transforming industries. It improves efficiency. It enables automation.','What is Deep Learning?']
for q in queries:
    print('\nQuery:',q)
    print('Response:',agent(q))
```
